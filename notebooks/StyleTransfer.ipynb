```python
from visualstoryteller.mixutils import load_content_image, load_local_image
from visualstoryteller.mixmorepics import GetStylePics
from tensorflow.nn import avg_pool
from matplotlib.pyplot import imshow, axis
```

```python
content_image = '/Users/ger/Desktop/marina.jpeg'
# False if it is a local file
content_web = False 

# style_image = '/Users/ger/Desktop/marina_pattern_1.jpg'
# style_image = '/Users/ger/Desktop/marina_pattern_2.jpeg'
style_image = '/Users/ger/Desktop/marina_pattern_3.jpeg'
# style_image = '/Users/ger/Desktop/marina_pattern_4.jpeg'
# style_image = '/Users/ger/Desktop/marina_blackwoman_style.jpeg'
# style_image = '/Users/ger/Desktop/marina_black_woman_style2.jpeg'

# False if it is a local file
style_web = False 

output_filename = 'marina_p3.jpg'

show = True # to show it or not
save = True # to save it or not
```

```python
#the class
mixing = GetStylePics()

#the content image
if content_web:
    mixing.content_image = [load_content_image(content_image)]
else:
    mixing.content_image = [load_local_image(content_image)]

# the style image
if style_web:
    mixing.style_image = [avg_pool(load_content_image(style_image), ksize=[3,3], strides=[1,1], padding='SAME')]
else:
    mixing.style_image = [avg_pool(load_local_image(style_image), ksize=[3,3], strides=[1,1], padding='SAME')]

mixing.stylize()

if save:
    mixing.save_jpgs(output_filename)
    
if show:
    imshow(mixing.stylized_image[0][0])
    axis('off');
```
